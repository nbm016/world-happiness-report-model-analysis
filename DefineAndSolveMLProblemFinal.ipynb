```python
filename_for_excel = os.path.join(os.getcwd(), "data_clustering", "WHR2018Chapter2OnlineData.xls")
df_regions = pd.read_excel(filename_for_excel, sheet_name='SupportingFactors')
```

```python
df_regions.columns
```

```python
# We see that there are too many countries to one hot encode. We will first retrieve regions for each country to see next steps.
# We saved access to an excel sheet of the WHR regions. We only want the countries and region indicators from this sheet. 

df_regions = df_regions[['country', 'Region indicator']]
df_regions
```

```python
df_regions.rename(columns={'Region indicator': 'Region'}, inplace=True)
```

```python
print(df_regions['Region'].nunique())
df_regions['Region'].value_counts()
```

```python
# We see that there are only 10 unique regions. 
# This is a better number of values to one hot encode compared to 150+ countries. 
# We will join df_regions with the main dataframe df_updated.

df_updated = df_updated.merge(df_regions, on='country')

df_updated.head()
```

```python
np.sum(df_updated.isnull(), axis=0)
```

```python
df_updated = df_updated.groupby(['country', 'Region']).mean().reset_index()
df_updated.head(15)
```

```python
# We combined the examples by country by taking the average of the feature values. 
# Examples for a country had additional rows for the different years. 
# After combining, we can drop the year column.

df_updated.drop(columns='year', inplace=True)

df_updated.columns
```

```python
# Now, we will one hot encode the Region column. 
# After doing this, the region and country columns will not be needed in the dataframe for modeling. 
df_regions_encoded = pd.get_dummies(df_updated['Region'], prefix='Region')
df_regions_encoded
```

```python
df_updated = df_updated.join(df_regions_encoded)
df_updated
```

```python
# Drop region and country from the df
df_updated.drop(columns=['country', 'Region'], inplace=True)
df_updated.columns
```

```python
print('Begin RF Implementation...')

rf_model = RandomForestRegressor(max_depth=32, n_estimators=100)
rf_model.fit(X_train, y_train)

print('End')
```

```python
# 1. Use the fitted model to make predictions on the test data
y_rf_pred = rf_model.predict(X_test)

# 2. Compute the RMSE
rf_rmse = mean_squared_error(y_test, y_rf_pred, squared=False)

# 3. Compute the R2 score
rf_r2 = r2_score(y_test, y_rf_pred)

print('[RF] Root Mean Squared Error: {0}'.format(rf_rmse))
print('[RF] R2: {0}'.format(rf_r2))
```
